## 5. Zmienianie wartości w ramkach danych

```python
# Chcemy zamienić miesiąc marzec `mar` na liczbę 3
# Spróbujmy w "zwykły" sposób

df_fire_copy = df_fire.copy(deep=True)

df_fire_copy[df_fire_copy['month'] == 'mar']['month'] = 3
```

```python
df_fire_copy
```

```python
# Wykorzystanie metody loc


df_fire_copy.loc[df_fire_copy['month'] == 'mar', 'month'] = 3
```

```python
df_fire_copy
```

```python
# Wykorzystanie metody apply
df_fire_copy = df_fire.copy(deep=True)
df_fire_copy['month'] = df_fire_copy['month'].apply(lambda x: 3 if x == 'mar' else x)
```

```python
df_fire_copy
```

```python
# Wykorzystanie metody map

df_fire_copy = df_fire.copy(deep=True)

map_dict_months = {
    'jan': 1,
    'feb': 2,
    'mar': 3,
    'apr': 4,
    'may': 5,
    'jun': 6,
    'jul': 7,
    'aug': 8,
    'sep': 9,
    'oct': 10,
    'nov': 11,
    'dec': 12,
}

map_dict_days = {
    'mon': 1,
    'tue': 2,
    'wed': 3,
    'thu': 4,
    'fri': 5,
    'sat': 6,
    'sun': 7,
}


df_fire_copy['month'] = df_fire_copy['month'].map(map_dict_months)
df_fire_copy['day'] = df_fire_copy['day'].map(map_dict_days)
```

```python
df_fire_copy
```
