```python
value = [df.iloc[11][1], df.iloc[11][3], df.iloc[11][5], df.iloc[11][7], df.iloc[11][9],
                      df.iloc[11][11]+df.iloc[11][13]]
names = ["Total","Biomass", "Metal ores", "Non-metallic minerals", "Fossil energy materials", "Other products"]
```

```python
data = {'value': [df.iloc[11][1], df.iloc[11][3], df.iloc[11][5], df.iloc[11][7], df.iloc[11][9],
                      df.iloc[11][11]+df.iloc[11][13]],
        'name': ["Total","Biomass", "Metal ores", "Non-metallic minerals", "Fossil energy materials", "Other products"]}
data = pd.DataFrame(data)
data
```

```python
percentage = pd.DataFrame(data['value']/data.iloc[0]['value']*100)
percentage = percentage.round(2)
percentage['value'] = percentage['value'].astype(str) + '%'
percentage
```

```python
data = data.sort_values('value', ascending = False)
percentage = [percentage.iloc[0]['value'], percentage.iloc[1]['value'],percentage.iloc[2]['value'],percentage.iloc[3]['value'],  percentage.iloc[4]['value'], percentage.iloc[5]['value']]
```

```python
plt.figure(figsize=(15, 11),facecolor="0.2")
fig, ax = plt.subplots(figsize=(15, 11),facecolor="0.2" )
plt.rcParams['axes.facecolor'] = '0.2'
bar = ax.bar(data['name'], data['value'], color = 'darkgreen')
label = plt.ylabel("consumption in tons")
label.set_color("white")
label1 = plt.xlabel("materials")
label1.set_color("white")
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
for i,rect in enumerate(bar):
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height+0.1,  height, ha='center', va='bottom')
    proc = percentage[i]
    plt.text(rect.get_x() + rect.get_width()/2.0, height-0.2, proc, ha='center', va='bottom')
plt.title("Domestic material consumption per person in the EU, 2019")


###### W związku z wcześniejszą uwagą Pani Anny Kozak, nie wprowadzam dodatkowego kolorowania każdego ze słupków na wykresie. Pani wykładowczyni uznała to za zabieg niepotrzebny,
###### zatem nie powielam tego błędu.
```
