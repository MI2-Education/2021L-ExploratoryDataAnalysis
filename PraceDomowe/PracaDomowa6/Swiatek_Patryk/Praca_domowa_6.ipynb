## Wykres poprawiony

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
labels = ["Jan-20", "Feb-20", "Mar-20", "Apr-20", "May-20", "Jun-20", "Jul-20", "Aug-20",
          "Sep-20", "Oct-20", "Nov-20", "Dec-20", "Jan-21", "Feb-21"]
wartosci = [361976, 381344, 428648, 286992, 382192, 423336, 416492, 413708, 409392, 387060, 
              315960, 247528, 277992, 287924]
breaks = np.linspace(0,450000, 10)
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]*1.01,y[i], ha='center', size=7)
```

```python
x = np.arange(len(labels))
width = 0.9  # szerokosc kolumn


fig, ax = plt.subplots(figsize=(8, 6), dpi=180)

plt.margins(x=0.015)

rects1 = ax.bar(x, wartosci, width)

ax.set_ylabel("National New Listings Count", fontweight="bold")
ax.set_xlabel("Month", fontweight="bold")
ax.set_title("Homes for sales supply in the housing market in each month", fontweight="bold")
plt.xticks(np.linspace(0,13,14))
ax.set_xticklabels(labels, size=7)
plt.yticks(breaks, size=8)

addlabels(labels, wartosci)

plt.show()
```

Dlaczego wykres jest lepszy niż oryginalny?

-	Oryginalny wykres jest w wersji 3d, przez co jest mniej czytelny, zaś mój wykres jest w wersji 2d, co ułatwia dostrzeganie i odpowiednią interpretację różnic w danych wartościach na wykresie.

-	Wartości na wykresie nie są w żaden sposób podzielone na kategorie (oprócz tych z osi X i Y), więc podział kolumn na różne kolory jest niepotrzebny, wręcz zbędny, bo łatwo może rozproszyć czytelnika. Ujednolicenie koloru kolumn ułatwia skupienie się na samych wartościach i ich tendencji, dzięki temu łatwiej i szybciej można wyciągać niezbędne wnioski.
