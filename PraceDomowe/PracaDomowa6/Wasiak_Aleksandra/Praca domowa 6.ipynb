```python
import pandas as pd
from matplotlib import pyplot as plt
```

**Oryginalna wizualizacja**

*źródło:* https://www.statista.com/chart/4530/twitter---facebook-15-years/

```python
dane = {"tekst": ["Daily Active Users (million)","Daily Active Users (million)",
                           "Revenue (billion $)","Revenue (billion $)",
                           "Net Profit (billion $)","Net Profit (billion $)",
                           "Employess (thousand)","Employess (thousand)",
                           "Market Capitalization (billion $)",
                           "Market Capitalization (billion $)"],
        "liczba": [192,1523,3.72,55.84,-1.14,22.11,5.5,35.587,53.3,483],
        "serwis": ["Twitter","Facebook","Twitter","Facebook","Twitter",
                            "Facebook","Twitter","Facebook","Twitter","Facebook"]
       }

DF = pd.DataFrame(dane)
```

```python
DF1 = DF[DF["tekst"] == "Daily Active Users (million)"]
DF2 = DF[DF["tekst"] == "Revenue (billion $)"]
DF3 = DF[DF["tekst"] == "Net Profit (billion $)"]
DF4 = DF[DF["tekst"] == "Employess (thousand)"]
DF5 = DF[DF["tekst"] == "Market Capitalization (billion $)"]
```

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1,5)

colormap = {
    "Facebook": "darkblue",
    "Twitter": "dodgerblue"
}

fig.set_figheight(5)
fig.set_figwidth(20)

fig.suptitle("Key metrics 15 years into the existence of Facebook/Twitter")

ax1.set_title("Daily Active Users (million)")
ax1.bar(DF1["serwis"], DF1["liczba"], color = DF1["serwis"].map(colormap))
ax2.set_title("Revenue (billion $)")
ax2.bar(DF2["serwis"], DF2["liczba"], color = DF2["serwis"].map(colormap))
ax3.set_title("Net Profit (billion $)")
ax3.axhline(0,color="grey")
ax3.bar(DF3["serwis"], DF3["liczba"], color = DF3["serwis"].map(colormap))
ax4.set_title("Employess (thousand)")
ax4.bar(DF4["serwis"], DF4["liczba"], color = DF4["serwis"].map(colormap))
ax5.set_title("Market Capitalization (billion $)")
ax5.bar(DF5["serwis"], DF5["liczba"], color = DF5["serwis"].map(colormap))

plt.show()
```

**Uzasadnienie:** Wysokości sąsiadujących słupków łatwiej jest porównać niż pola sąsiadujących kół.
