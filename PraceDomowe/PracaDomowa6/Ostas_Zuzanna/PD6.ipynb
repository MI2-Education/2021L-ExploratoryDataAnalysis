### Kod poprawionego wykresu

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = {
    "Program": ["Dance Dance", "Twoja Twarz Brzmi Znajomo"],
    "Średnio": [1.823, 1.543],
    "Pik": [2.023, 1.865]
}

df = pd.DataFrame(data)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)

fig.set_figheight(5)
fig.set_figwidth(15)

fig.suptitle('Porównanie oglądalności dwóch programów', size=15, x=0.5, y=0.99)


ax1.bar(df['Program'], df['Średnio'], 0.7)
ax2.bar(df['Program'], df['Pik'], 0.7)


ax1.set_title('Średnia oglądalność')
ax1.set_xlabel('Program')
ax1.set_ylabel('Oglądalność w mln')

ax2.set_title('Pikowa oglądalność')
ax2.set_xlabel('Program')
ax2.set_ylabel('Oglądalność w mln')


rects1 = ax1.bar(df['Program'], df['Średnio'])
rects2 = ax2.bar(df['Program'], df['Pik'])

ax1.bar_label(rects1, padding=3)
ax2.bar_label(rects2, padding=3)

plt.style.use("seaborn")
```

### Uzasadnienie

Na poprawionym wykresie słupki są pokazanie w 2D zamiast w 3D, co pozwala na lepsze porównywanie
wysokości słupków przez odbiorcę. Porównanie ułatwiają też linie siatki w tle. Dodatkowo, zakładając, że wykres miał porównywać dwa programy, a nie średnią i pikową oglądalność dla tych programów, w nowym wykresie zmienione jest ułożenie poszczególnych kolumn. Po lewej stoją kolumny obu programów dla średniej oglądalności, a po prawej dla pikowej
oglądalności. Dzięki temu możemy łatwiej porównać przedstawione na wykresie programy.
