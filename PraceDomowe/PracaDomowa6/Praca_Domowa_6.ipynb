## Wizualizacja 

```python
import pandas as pd

sondaz = {'procenty': [6.0, 6.2, 7.4, 12.1, 17.9, 18.1, 32.3],
        'nazwy': ['PSL - Koalicja Polska','Konfederacja','Lewica',
'Nie wiem/Trudno powiedzieć','Polska 2050', 'Koalicja Obywatelska','Prawo i Sprawiedliwość'] }

sondaz = pd.DataFrame(sondaz, columns = ['nazwy', 'procenty'])

print(sondaz)
```

```python
fig, ax = plt.subplots(figsize=(10,7))

plt.barh(sondaz['nazwy'],sondaz['procenty'], color = "#B0B0B0")

plt.suptitle("Na które ugrupowanie zagłosował(a)by \n Pan(i) w wyborach do Sejmu?", y=1.02, x = 0.45, 
             fontsize=20, weight = 'bold')
plt.title("badanie przeprowadzone przez IBRiS w dniach 05.03-06.03.2021", x = 0.68, fontsize=11)
ax.set_xlabel("procent ankietowanych", fontsize=11)

for i, v in enumerate(sondaz['procenty']):
    ax.text(v - 2.5, i, str(v), color='black', fontsize=12, ha='left', va='center')

ax.grid(True)
ax.set_xlim(0, 35) 
plt.show()
```
