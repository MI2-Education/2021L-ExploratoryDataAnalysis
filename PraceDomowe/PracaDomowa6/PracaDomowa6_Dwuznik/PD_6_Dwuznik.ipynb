Wykres pochodzi ze strony Wirtualny Nowy Przemysł i ma obrazować wyniki sondażu parlamentarnego z dnia 15.03.2021 przeprowadzonego przez Kantar. O ile same wysokości słupków i przedstawione dane nie są przekłamane, to wykres zawiera parę błędów, które wpływają na jego odbiór.

Mylić może fakt, że nie są opisane linie pomocnicze, przyglądając się wynikom Lewicy, KO i PiS możemy dojść do wniosku, że jedna kreska ma oznaczać około 9%, co może być mylące. Bez spojrzenia na wartości byłem intuicyjnie przekonany, że odstępy są 10 – procentowe. 
Wykres nie jest fatalny, ale jest raczej prostą wizualizacją do artykułu, niż porządną reprezentacją danych, więc możemy zrobić to lepiej!

## 2.Propozycja poprawy wykresu

```python
import pandas as pd
import matplotlib.pyplot as plt

dane= {'wyniki': [2, 3, 5, 8, 14, 24, 30],
        'partie': ['Kukiz15','Polskie Stronnictwo Ludowe', 'Konfederacja','Lewica',
'Polska 2050', 'Koalicja Obywatelska','Prawo i Sprawiedliwość'] }

ramka = pd.DataFrame(dane, columns = ['partie', 'wyniki'])

plt.figure(figsize=(10, 6))
plt.bar(ramka['partie'], ramka['wyniki'], color='gray')

plt.xticks(rotation='vertical')
plt.suptitle("Sondaż poparcia dla partii politycznych", y=1.01, x=0.40,
             fontsize=16, weight = 'bold')
plt.title("Kantar - 15.03.2021",y= 1.02, x =0.15)
plt.grid(color = 'grey')
plt.ylabel('procent poparcia', fontsize=12)
plt.show()
```

Stworzony przeze mnie wykres, można jednoznacznie rozumieć bez uprzedniego zapoznania się z artykułem. Bliżej nieokreśloną wcześniej skalę zastąpiła pomocna przy odczytaniu wartości siatka, z odstępami między kolejnymi liniami odpowiadającymi 5 punktom procentowym. Głównym elementem wykresu są odpowiednio oznaczone słupki, co czyni go znacznie czytelniejszym.
