### Nowy wykres

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

data = {
    "rok": [2013,2014,2015,2016,2017,2018,2019,2020],
    "mld": [42.3, 39.7, 41.2, 34.4, 25.7, 19, 17.7,0],
    "procenty": [27, 24.4, 24.7, 20.4, 14.3, 9.9, 9.7,8.45]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8.5,5.5))

plt.bar(df['rok'], df['mld'], color = 'slateblue')
for i in range(7):
    plt.text(x = df.rok[i]-0.25, y = df.mld[i]+0.5, s = df.mld[i],
            fontdict = dict(color = 'darkblue', size = 8, weight = 'bold'))
plt.xticks([2013,2014,2015,2016,2017,2018,2019,2020]) 
    
plt.title('Ile wynosiła luka w VAT?', fontsize = 12, color = 'darkred',fontweight='bold')
plt.xlabel('Rok',fontsize = 10, color = 'darkred')
plt.ylabel('Nominalnie w mld zł', fontsize = 10, color = 'darkred')

plt.scatter(df['rok'], df['procenty'], c = 'gold', s = 50, marker = "D", zorder = 10)
plt.scatter(2020.95, 24.8, c = 'gold', s = 45, marker = "D")

procenty_napisy = ["27%","24,4%","24,7%","20,4%","14,3%","9,9%","9,7%","8,2-8,7%"]

for i in range(8):
    plt.text(x = df.rok[i]-0.3, y = df.procenty[i]+1.2, s = procenty_napisy[i], 
          fontdict = dict(color = 'gold', size = 8, weight = 'bold'))

plt.text(x = 2020-0.15, y = 1, s = 'BD', 
         fontdict = dict(color = 'darkblue', size = 8, weight = 'bold'))
plt.text(x = 2021, y = 0.5, s = 'Źródło: Fundacja CASE,\n Ministerstwo Finansów', 
         fontdict = dict(color = 'black', size = 6))
plt.text(x = 2020.6, y = 7, s = '*dla 2020 wstępne, \n przybliżone rachunki', 
         fontdict = dict(color = 'black', size = 6.5))
plt.text(x = 2020.7, y = 23.2, s = '            - Jako procent \n              teoretycznych wpływów', 
          fontdict = dict(color = 'black', size = 6.5),
         bbox = dict(facecolor = 'w', alpha = 0.3))
sns.despine()
plt.show()
```

Na wykresie postanowiłam zmienić nastepujące elementy:
1. **oznaczenie osi** - na początkowej wizualizacji brak było podpisanej osi y, natomiast na osi x były
tylko niektóre wartości, a niektóre znalazły się na słupkach, lepiej to wygląda gdy jest ujednolicone,
2. **wymiar słupków** - jak wiadomo, słupki 2D są bardziej czytelne niż słupki 3D, nie tworzą zbędnej
perspektywy, która może zaburzać prawidłowe odczytywanie danych,
3. **kolor i łączenie punktów** - czerwone punkty oraz linia na początkowym wykresie mogłyby sugerować,
że spadek danych wartości procentowych jest zjawiskiem negatywnym, jednak w rzeczywistości im
mniejszy procent wpływów stanowi luka VAT, tym lepiej, dlatego użyłam bardziej neutralnego koloru,
jakim jest żółty. Nie ma także zbytnio potrzeby łączenia tych punktów, gdyż każda procentowa wartość
jest jedna i stała dla danego roku, a linia ta, moim zdaniem, niekorzystnie wpływa na estetykę wykresu.
