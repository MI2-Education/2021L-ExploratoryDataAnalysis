## Kod i propozycja wizualizacji

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
   
Sezon = [1, 2, 3, 4, 5]
Ocena = [8.79, 8.86, 8.81, 9.02, 9.43]

plt.bar(Sezon, Ocena, color = "seagreen", zorder = 2)
plt.grid(axis = 'y')
plt.title('Ocena sezonów serialu Breaking Bad', size = 20)
plt.ylabel('Ocena', size = 20)
plt.rcParams.update({'font.size': 20})
plt.xlabel('Sezon', size = 20)

for i in range(5):
    plt.text(Sezon[i] - 0.15, Ocena[i]+0.1, Ocena[i], fontdict = dict(color = 'black', size = 17))

plt.show()
```
