## Błędy pojawiające się na wykresie

1. Wartości procentowe opisujące oś OY są przesunięte względem odpowiadającym im poziomym liniom, biegnącym wzdłuż wykresu.
2. Kolumny wykresu są w 3D, przez co występują wizualne zniekształcenia i problemy w odczytaniu danych (przykładowo można odnieść wrażenie, że kolumna z 2019r jest wyższa niż kolumna z 2018 roku, pomimo że odpowiadająca jej wartość procentowa jest mniejsza)
3. W stosunku do największej wartości procentowej luki Vat osiągniętej w 2013 roku (25,7%) nastąpił spadek tej luki do 12 % w 2019 roku, czyli o 13.7 p.p., a nie 12.3 p.p., jak jest to zaprezentowane na wykresie (ponad niebieską strzałką)

## Poprawiona wersja

```python
import matplotlib.pyplot as plt
import pandas as pd
Dane = pd.DataFrame({'Lata':['2008', '2009', '2010','2011','2012','2013','2014','2015','2016','2017','2018','2019'], 
                     'Procent':[15.5, 20.3,18.7,19.7,25.6,25.7,24.3,24.2,20.2,14.6,12.1,12]})
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
for i in range(1,6):
    plt.axhline(y=5*i,color = "grey", linestyle ="--", linewidth = 0.5)
ax.bar(Dane["Lata"],Dane["Procent"], color = ["orange","orange","orange","orange","orange","orange","orange","orange","blue","blue","blue","blue"],  )
plt.xlabel('Rok')
plt.ylabel('Procent')
plt.title(' Luka Vat')




def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha = 'center') # funckja, która przypisuje każdej kolumnie wykresu label z wartością i wyświetla go ponad daną kolumną.

addlabels(Dane["Lata"], Dane["Procent"])
plt.show()
```

## Opis poprawionej wersji

1. Oś OY jest opisana jasno i przejrzyście, wartości procentów przy osi OY są zgodne z horyzontalnymi liniami.
2. Kolumny są zrobione w 2D, dzięki czemu łatwo porównać wysokość sąsiednich kolumn i nie ma zaburzeń w odbiorze wizualnym wykresu.
3. Pozbywamy się błędnego napisu opisującego spadek Luki Vat o 12.3 p.p.
