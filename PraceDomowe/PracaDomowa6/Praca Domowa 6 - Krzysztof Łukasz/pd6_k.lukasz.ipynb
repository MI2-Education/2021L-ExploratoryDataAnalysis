
## Omawiany wykres  

W celu zilustrowania sukcesu Narodowego Programu Szczepień w Wiadomościach pojawił się taki wykres:



Niestety, wygląda to na dobieranie danych pod określoną tezę. Spróbujmy wykonać podobny wykres, bazując na większym zbiorze danych. Użyte dane pochodzą ze strony `https://ourworldindata.org/covid-vaccinations`  i zawierają informacje o liczbie wykorzystanych dawek szczepionek z 66 państw, stan na 22lutego 2021.


```python
Image('wykres.png')
```

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(20, 10))
colors = {'Pokazane w TVP':'red', 'Polska':'green', 'Pozostałe kraje':'skyblue'}
plt.bar("Country", "Applied doses per 100", data = df,  color=df['Grupa'].map(colors))
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels)
plt.xlabel("Kraj")
plt.xticks(rotation = 90)
plt.ylabel("Wykorzystane dawki")
plt.title("Podane dawki szczepionki na 100 mieszkańców")
plt.show()
```

Wykres ten jest lepszy od oryginalnego, ponieważ pokazuje pełniejszy obraz rzeczywistej sytuacji ze szczepionkami w różnych krajach, w przeciwieństwie do wykresu z Wiadomości, który został wykonany, aby pasował do postawionej wcześniej tezy. Ponadto, nieuwzględnienie państw o wyższym wskaźniku wyszczepienia populacji powoduje wyolbrzymienie różnic między krajami, co nie ma miejsca na drugim wykresie.
