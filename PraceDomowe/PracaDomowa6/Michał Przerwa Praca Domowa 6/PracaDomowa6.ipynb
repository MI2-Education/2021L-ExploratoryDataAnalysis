# Wykres do poprawienia

Poprawiany wykres w pliku 'wykres_zly.png'

Wykres jest przedstawiony w formie kołowej, utworzymy wykres słupkowy.

```python
import pandas as pd

sejm = {'Partia': ['PiS','Polska 2050','KO','Konfederacja','Lewica','Mniejszość niemiecka'],
        'Liczba_miejsc': [166,136,84,42,31,1]
        }

df = pd.DataFrame(sejm, columns = ['Partia', 'Liczba_miejsc'])
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plot = sns.barplot(x="Partia", y="Liczba_miejsc", color="lightblue", data=df)

for i, p in enumerate(plot.patches):
    height = p.get_height()
    plot.text(p.get_x()+p.get_width()/2., height + 1, df['Liczba_miejsc'][i], ha="center")
    
plot = plot.set_xticklabels(plot.get_xticklabels(),rotation=90)
plot = plt.ylabel('Liczba miejsc w Sejmie')
plot = plt.title('Prognozowana liczba miejsc w Sejmie RP na nastepna kadencję')
```

## Dlaczego ten wykres jest lepszy?

a) wykres słupkowy jest łatwiejszy w interpretacji

b) dobrze opisane osie przedstawiają zarówno nazwy partii, jak i liczbę miejsc, która jest dodatkowo podkreśloną nad słupkami
