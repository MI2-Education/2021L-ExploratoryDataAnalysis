# Poprawiony wykres
## Metoda

```python
dane_tabela_ekstra_pensja = {
    "Stopień" : ["I 5/5 + 1/1", 'II 5/5 + 0/1', 'III 4/5 + 1/1', "IV 4/5 + 0/1",
             "V 3/5 + 1/1", "VI 3/5 + 0/1", "VII 2/5 + 1/1", "VIII 2/5 + 0/1"],
    "Graczy" : [0, 1, 27, 89, 605, 1645, 5221, 14642],
    "Kwota w zł" : [5000, 25000, 1000, 200, 80, 25, 10, 5]
}
dane_tabela_wykres = {
    "Stopień wygranej": ["I", 'II', 'III', "IV",
             "V", "VI", "VII", "VIII"],
    "Łączna kwota" : [0,0,0,0,0,0,0,0]
}
tabela_ekstra_pensja = pd.DataFrame(dane_tabela_ekstra_pensja)
tabela_wykres = pd.DataFrame(dane_tabela_wykres)
tabela_wykres["Łączna kwota"] = tabela_ekstra_pensja["Graczy"] * tabela_ekstra_pensja["Kwota w zł"]
```

## Efekt końcowy

```python
def add_labels(x,y):
     for i in range(len(x)):
            plt.text(i,y[i] +y.max()/100,y[i], ha= "center")
#funkcja dodaje napisy wartości nad słupki: y[i] +y.max()/100 odpowiada za umieszczenie na odpowiedniej wysokości nad słupkiem

plt.figure(figsize=(12, 8))
plt.bar(tabela_wykres["Stopień wygranej"], tabela_wykres["Łączna kwota"], tick_label = tabela_wykres["Stopień wygranej"])
add_labels(tabela_wykres["Stopień wygranej"], tabela_wykres["Łączna kwota"])
plt.title('Łączne kwoty poszczególnych stopni wygranych w losowaniu Ekstra Pensja z 19.03.2021')
plt.xlabel("Stopień wygranej")
plt.ylabel("Łączna kwota w złotówkach")

plt.show()
```

# Podsumowanie
Przedstawiony wykres jest zdecydowanie lepszy od opublikowanego, ponieważ jako wykres słupkowy znacznie lepiej oddaje różnice pomiędzy kwotami wygranymi w ramach kolejnych stopni - nawet bez podania kwot nad słupkami bardzo łatwo można by porównać każdy ze stopni z dowolnym innym. Dodatkowo, nie istnieje problem tajemniczego zniknięcia 1 pkt procentowego (który zapewne zaginął podczas zaokrąglania).
