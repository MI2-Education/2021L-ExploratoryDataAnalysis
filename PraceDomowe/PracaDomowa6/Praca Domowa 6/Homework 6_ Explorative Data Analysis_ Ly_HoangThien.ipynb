# Downside of this visualization

This visualization goes totally wrong from the first eye's glance by the following reasons:
1. Having mistake in term of using annotation. The pie in this 3-D piechart for 39.8% is much bigger than the pie for 60.2%.
2. Misleading the color! The bigger pie should be colored in red, and vice versa for the smaller pie should be in green colour.
3. The superflousness of using 3-D pie chart, because the 3-D chart sometimes makes us hard to estimate and give a precise judgement. Even, the chart is worsen in term of the first pie did not start at the place of 0:00 hour. It creates more obstacles for readers/audiences to understand what this chart is about.

# Corrected chart

I will suggest two options to correct the above chart:

*Option 1*: use a 2-D pie chart with appropriate colors and annotations.

*Option 2*: (better): use a bar chart reversed horizontally.

## Code of first proposed visualization:


```python
my_data= [60.2,39.8]
my_labels="No","Yes"

# Choosing colors:
my_colors=['#ff0000',"#33eb00"]

# Explode the "Yes":
my_explode = (0, 0.05)

# Creating 2-D pie chart with data and labels, colors, explode, autopct gives annotations, startangle: starting angle
patches, texts, autotexts = plt.pie(my_data,labels=my_labels,colors=my_colors,
                                    explode=my_explode,autopct='%1.1f%%',startangle=90)

# Adding title
plt.title("Do you think Indian Govt has taken suitable steps to give China a befitting reply?")

# With equal axis
plt.axis('equal')


# Changing font size of annotation:

texts[0].set_fontsize(15)
texts[1].set_fontsize(15)
autotexts[0].set_fontsize(14)
autotexts[1].set_fontsize(14)

plt.tight_layout()
plt.show()
```

## Code of the second proposed visualization:



```python
my_data= [60.2,39.8]
my_labels=["No","Yes"]


# Using the style of ggplot
plt.style.use('ggplot')


x=plt.barh(my_labels,my_data,color=['#ff0000',"#33eb00"])
plt.title('Do you think Indian Govt has taken suitable steps to give China a befitting reply?')
plt.xlabel('Percent')
plt.ylabel('Answers')


plt.show()
```

# Overview:

An arbitrary chart from two upon proposed charts is better than the 3-D pie charts published on the TV-news, as 2D- pie chart with the correct use of colors (complies with our intuition) and annotations is much better than the 3-D pie chart with wrong colors and wrong annotations. Moreover, it is much much superior for the horizontal pie chart, since readers can quickly compare between each answers and draw a comprehensive overview immediately.
