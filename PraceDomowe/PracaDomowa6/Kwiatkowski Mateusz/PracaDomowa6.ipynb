## Wady przedstawionego wykresu
1. Największą wadą tej wizualizacji jest nieodpowiednia forma wykresu, która prowadzi do złej interpretacji zaprezentowanych danych. Zastosowanie wykresu liniowego sugeruje, że przedstawione dane zmieniają się w sposób ciągły, co jest nieprawdą. Wizualizacja miała pokazywać jak rozkładało się posiadanie piłki między zespołami w poszczególnych piętnastominutowych przedziałach. Posiadanie piłki w dwóch różnych przedziałach było liczone osobno, więc dane te są od siebie niezależne. Połączenie punktów jest zatem zupełnie nieuzasadnione.
2. Kolejną wadą, która uniemożliwia szybkie i prawidłowe odczytanie danych z wykresu jest brak legendy. Autor zapomniał podpisać, którymi kolorami oznaczone są zespoły. Również w tekście zamieszczonym na stronie nie pojawiła się taka informacja. By zidentyfikować, który kolor pokazuje posiadanie piłki przez Legię, musiałem skorzystać z innych źródeł.
3. Autor wykresu nie dodał skali na osi y. Jak można zauważyć, nie zaczyna się ona od 0.
4. Przedziały opisane na osi x zachodzą na siebie. Nie wiadomo, do którego przedziału zostało wliczone posiadanie piłki z 15-stej minuty.
5. Oczywiste jest, że udział w posiadaniu piłki przez oba zespoły zawsze sumuje się do 100% (ponieważ czas posiadania liczy się do momentu przejęcia piłki przez drużynę przeciwną). Wykresy dla obu zespołów muszą być zatem symetryczne względem poziomu 50%. Autor mógł skorzystać z tego faktu i przedstawić na wykresie dane dla jednego zespołu.

## Mój wykres

```python
from matplotlib import pyplot as plt
x=["0-15","16-30","31-45","46-60","61-75","76-90"]
y=[61, 57, 41, 50, 41, 53]
plt.rcParams["figure.figsize"] = (9,7)
plt.bar(x, y)
plt.title('Posiadanie piłki przez Legię w meczu z Zagłębiem Lubin')
plt.xlabel('Przedziały czasowe (w minutach)')
plt.ylabel('Udział w posiadaniu piłki (w procentach)')
plt.ylim((0,67))
for i in range(6):
    plt.text(i-.15, y[i] + .5, str(y[i])+"%")
plt.show()
```

## Wprowadzone modyfikacje
Najważniejszą zmianą, jaką wprowadziłem, jest zmienienie formy wizualizacji na wykres kolumnowy. Taka forma pokazuje, że dane w każdym kwadransie liczone były osobno, co (w odróżnieniu od oryginalnego wykresu) nie wprowadza odbiorców w błąd. Na moim wykresie ograniczyłem się do zaprezentowania posiadania piłki przez Legię (ponieważ serwis legia.net skierowany jest głównie do kibiców tego zespołu). Dane dotyczące Zagłębia pominąłem dla poprawienia czytelności wizualizacji, jednak wciąż można je wywnioskować z wykresu. Opisałem oś y oraz zmieniłem jej skalę, by zaczynała się od 0. Dzięki temu proporcje między wysokościami kolumn są poprawne. Pozostawiłem na wykresie dane procentowe, żeby odbiorcy mogli dokładnie odczytać dane. Wprowadziłem również drobną modyfikację na osi x. Zmieniłem przedziały czasowe, tak by nie nachodziły na siebie.
