## Kod

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('dane.csv')

colormap = {
    'Video': 'red',
    'News': 'yellow',
    'Audio': 'green',
    'Video/Audio': 'orange'
}

plt.figure(figsize=(15, 10))

plt.bar(df['Service'], df['Subscribers'], color=df['Type'].map(colormap))
plt.xticks(rotation=90)

plt.title('Wykres liczby subskrybentów serwisów streamingowych')
plt.xlabel('Serwis streamingowy')
plt.ylabel('Liczba subskrybentów w mln')

labels = list(colormap.keys())
handles = [plt.Rectangle((0,0),1,1, color=colormap[label]) for label in labels]
plt.legend(handles, labels)


plt.show()
```
