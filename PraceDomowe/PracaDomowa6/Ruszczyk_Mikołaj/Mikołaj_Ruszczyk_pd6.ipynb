```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = {'2019': [2124118,2187680,2401899,2460560,3280645,3605839,4059520,4319590,3347843,2987993,2596441,2295963],
        '2020': [2291506,2378371,881365,86386,387177,1314252,2711895,3208081,2194007,1363224,586655,476050]}
df = pd.DataFrame(data,columns=['2019','2020'], index = ['styczen','luty','marzec','kwiecien','maj','czerwiec',
                                                         'lipiec','sierpien','wrzesien','pazdziernik','listopad','grudzien'])
```

```python


df.plot.barh()
plt.gca().invert_yaxis()
plt.title("Liczba turystów korzystających z noclegów w 2019 i 2020")
plt.ylabel('Miesiąc')
plt.xlabel('')
plt.show()
```
