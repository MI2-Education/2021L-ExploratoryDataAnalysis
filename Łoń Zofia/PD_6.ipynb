## Kod i wykres

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

x = [12.82,8.36,7.10,7.10,6.32, 5.97,5.86,5.79,5.73]
kraje =  ["Malta","Dania","Rumunia","Polska","Hiszpania","średnia UE","Niemcy","Włochy","Francja"]

plt.rcParams["figure.figsize"]=12,8
plt.bar(kraje,x, align = 'center',color ='grey')
plt.title("Podane dawki szczepionki na 100 mieszkanców w krajach UE\nstan na 22 lutego")

for i in range(len(x)): #ten fragmet kodu umożliwił mi stworzenie 'etykiety' dla każdego słupka co zwiekszyło czytelność wykresu
    plt.annotate(str(x[i]), xy=(kraje[i],x[i]), ha='center', va='bottom') #funkcja annotate tworzy napis w odanym miejscu, ha i va odpowiadają za polozenie tekstu

plt.show()
```

## Uzasadnienie

Stworzony przeze mnie wykres przedstawia dane zgodnie z rzeczywistym stanem rzeczy (w przeciwieństwie do oryginalnego) i nie wprowadza zbędnego koloru zielonego (który tylko wzmaciał manipulacje odbiorcy jakoby Polska była liderem w tempie szczepień)
